```python
##### import classy module
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
from classy import Class
```

# LCDM

```python
def lcdm_background():
    # create instance of the class "Class"
    Lambda_cdm = Class()
    # pass input parameters
    Lambda_cdm.set({'omega_b':0.0223828,'omega_cdm':0.1201075,'h':0.67810,'A_s':2.100549e-09,'n_s':0.9660499,'tau_reio':0.05430842,'Omega_fld':0.,'Omega_scf':0.,'Omega_anq':0.,'Omega_sde':0.,'Omega_a2f':0.,'Omega_eymh':0.,'Omega_avf':0.})
    #Lambda_CDM.set({'output':'tCl,pCl,lCl,mPk','lensing':'yes','P_k_max_1/Mpc':3.0})
    # run class
    Lambda_cdm.compute()
    return pd.DataFrame(Lambda_cdm.get_background())
```

# Anisotropic quintessence

```python
def anq(l,mu):
    # create instance of the class "Class"
    ani_quin = Class()
    # pass input parameters
    ani_quin.set({'omega_b':0.0223828,'omega_cdm':0.1201075,'h':0.67810,'A_s':2.100549e-09,'n_s':0.9660499,'tau_reio':0.05430842,'Omega_fld':0.,'Omega_scf':0.,'Omega_sde':0.,'Omega_a2f':0,'Omega_eymh':0,'Omega_avf':0.,'Omega_anq':1.,'anq_parameters_1':l,'anq_parameters_2':mu})
    #LambdaCDM.set({'output':'tCl,pCl,lCl,mPk','lensing':'yes','P_k_max_1/Mpc':3.0})
    # run class
    ani_quin.compute()
    return pd.DataFrame(ani_quin.get_background())
```

```python
pd.DataFrame(anq(0.1,1.e3))
```

```python
z = anq(0.1,1.e3)['z']
H = anq(0.1,1.e3)['H [1/Mpc]']


Omega_r   = (anq(0.1,1.e3)['(.)rho_g']+anq(0.1,1.e3)['(.)rho_ur']) /H**2
Omega_b   = (anq(0.1,1.e3)['(.)rho_b'])/H**2
Omega_cdm = (anq(0.1,1.e3)['(.)rho_cdm'])/H**2
Omega_DE  = (anq(0.1,1.e3)['(.)rho_anq'])/H**2
w_DE      = (anq(0.1,1.e3)['(.)p_anq'])/anq(0.1,1.e3)['(.)rho_anq']



x_anq = anq(0.1,1.e3)['x_anq']/H
y_anq = anq(0.1,1.e3)['y_anq']/H
z_anq = anq(0.1,1.e3)['z_anq']/H

def Sigma(l,mu):
  return anq(l,mu)['sigma_anq']/anq(l,mu)['H [1/Mpc]']

def log_Sigma(l,mu):
  return math.log(anq(l,mu)['sigma_anq']/anq(l,mu)['H [1/Mpc]'])    

def w_DE(l,mu):
  return anq(l,mu)['(.)p_anq']/anq(l,mu)['(.)rho_anq']

q = -1. + (3.*x_anq*x_anq + 2.*z_anq*z_anq) + 3.*(Omega_b+Omega_cdm)/2. + 2.*Omega_r + 3.*Sigma(0.1,1.e3)**2 
w_eff = -1. + 2.*(1.+q)/(1-Sigma(0.1,1.e3)**2)/3 - 2.*Sigma(0.1,1.e3)**2/(1-Sigma(0.1,1.e3)**2)

```

```python
def solid(n,m):
    # create instance of the class "Class"
    solid1 = Class()
    # pass input parameters
    solid1.set({'omega_b':0.0223828,'omega_cdm':0.1201075,'h':0.67810,'A_s':2.100549e-09,'n_s':0.9660499,'tau_reio':0.05430842,'Omega_fld':0.,'Omega_scf':0.,'Omega_anq':0.,'Omega_a2f':0.,'Omega_eymh':0.,'Omega_avf':0.,'Omega_sde':1.,'sde_parameters_1':n,'sde_parameters_2':m})
    #LambdaCDM.set({'output':'tCl,pCl,lCl,mPk','lensing':'yes','P_k_max_1/Mpc':3.0})
    # run class
    solid1.compute()
    return pd.DataFrame(solid1.get_background())
```

```python
z = solid(0.01,0.00001)['z']
H = solid(0.01,0.00001)['H [1/Mpc]']

Omega_r   = (solid(0.01,0.00001)['(.)rho_g']+solid(0.01,0.00001)['(.)rho_ur'])/H**2
Omega_b   = (solid(0.01,0.00001)['(.)rho_b'])/H**2
Omega_cdm = (solid(0.01,0.00001)['(.)rho_cdm'])/H**2
Omega_DE  = (solid(0.01,0.00001)['(.)rho_sde'])/H**2
w_DE      = (solid(0.01,0.00001)['(.)p_sde'])/solid(0.01,0.00001)['(.)rho_sde']

f1_sde = (solid(0.01,0.00001)['f1_sde'])/H
f2_sde = (solid(0.01,0.00001)['f2_sde'])/H

def Sigma(n,m):
  return solid(n,m)['sigma_sde']/solid(n,m)['H [1/Mpc]']

q = (1. + (2*0.01 - 3)*f1_sde**2 + 2*(2*0.00001 - 3)*f2_sde**2 + Omega_r + 3*Sigma(0.01,0.00001)**2)/2.
w_eff = (2*q - 1)/3.

```

# Anisotropic 2-Form (a2f)

```python
def a2f(l,m):
    # create instance of the class "Class"
    a2f1 = Class()
    # pass input parameters
    a2f1.set({'omega_b':0.0223828,'omega_cdm':0.1201075,'h':0.67810,'A_s':2.100549e-09,'n_s':0.9660499,'tau_reio':0.05430842,'Omega_fld':0.,'Omega_scf':0.,'Omega_anq':0.,'Omega_sde':0.,'Omega_eymh':0.,'Omega_avf':0.,'Omega_a2f':1.,'a2f_parameters_1':l,'a2f_parameters_2':m})
    #LambdaCDM.set({'output':'tCl,pCl,lCl,mPk','lensing':'yes','P_k_max_1/Mpc':3.0})
    # run class
    a2f1.compute()
    return pd.DataFrame(a2f1.get_background())
```

```python
z = a2f(2,30)['z']
H = a2f(2,30)['H [1/Mpc]']

Omega_r   = (a2f(2,30)['(.)rho_g']+a2f(2,30)['(.)rho_ur']) /H**2
Omega_b   = (a2f(2,30)['(.)rho_b'])/H**2
Omega_cdm = (a2f(2,30)['(.)rho_cdm'])/H**2
Omega_DE  = (a2f(2,30)['(.)rho_a2f'])/H**2

x1_a2f = a2f(2,30)['x1_a2f']/H
x2_a2f = (a2f(2,30)['x2_a2f'])/H
OmB_a2f = (a2f(2,30)['OmB_a2f'])/H**2

def Sigma(l,m):
  return a2f(l,m)['sigma_a2f']/a2f(l,m)['H [1/Mpc]']

def w_DE(l,m):
  return a2f(l,m)['(.)p_a2f']/a2f(l,m)['(.)rho_a2f']

def w_eff(l,m): 
  return x1_a2f**2 - x2_a2f**2 + Sigma(l,m)**2 + Omega_r/3. - OmB_a2f/3.
```

```python
def EYMH(a,w):
    eymh = Class()
    eymh.set({'omega_b':0.0223828,'omega_cdm':0.1201075,'h':0.67810,'A_s':2.100549e-09,'n_s':0.9660499,'tau_reio':0.05430842,'Omega_fld':0.,'Omega_scf':0.,'Omega_anq':0.,'Omega_sde':0.,'Omega_a2f':0.,'Omega_sde':0.,'Omega_anq':0.,'Omega_a2f':0.,'Omega_avf':0.,'Omega_eymh':1.,'eymh_parameters_1':a,'eymh_parameters_2':w})
    eymh.compute()
    return pd.DataFrame(eymh.get_background())
```

```python
z = EYMH(1,-39.1439)['z']
H = EYMH(1,-39.1439)['H [1/Mpc]']

Omega_r   = (EYMH(1,-39.1439)['(.)rho_g']+EYMH(1,-39.1439)['(.)rho_ur']) /H**2
Omega_b   = (EYMH(1,-39.1439)['(.)rho_b'])/H**2
Omega_cdm = (EYMH(1,-39.1439)['(.)rho_cdm'])/H**2
Omega_DE  = (EYMH(1,-39.1439)['(.)rho_eymh'])/H**2

x_eymh   = EYMH(1,-39.1439)['x_eymh']/H
y_eymh   = (EYMH(1,-39.1439)['y_eymh'])/H
xi_eymh  = EYMH(1,-39.1439)['xi_eymh']
pp_eymh  = (EYMH(1,-39.1439)['pp_eymh'])/np.sqrt(H)
s_eymh  = (EYMH(1,-39.1439)['s_eymh'])/np.sqrt(H)
z_eymh   = (EYMH(1,-39.1439)['z_eymh'])/H
w_eymh  = EYMH(1,-39.1439)['w_eymh']/H
v_eymh   = (EYMH(1,-39.1439)['v_eymh'])/H

def Sigma(a,w):
  return EYMH(a,w)['sigma_eymh']/EYMH(a,w)['H [1/Mpc]']

def w_DE(a,w):
  return EYMH(a,w)['(.)p_eymh']/EYMH(a,w)['(.)rho_eymh']

q = 1. - 2.*v_eymh*v_eymh- w_eymh*w_eymh + z_eymh*z_eymh - 0.5*(Omega_b+Omega_cdm) + Sigma(1,-39.1439)**2

w_eff = (2*q - 1)/3 


psi_1  = np.sqrt(3)*(EYMH(1,-34.5388)['pp_eymh'])/EYMH(1,-34.5388)['s_eymh']/EYMH(1,-34.5388)['xi_eymh']
psi_2  = np.sqrt(3)/EYMH(1,-34.5388)['xi_eymh']
dpsi_1 = psi_1*(2*Sigma(1,-34.5388) - 1) + np.sqrt(3)* EYMH(1,-34.5388)['x_eymh']/EYMH(1,-34.5388)['H [1/Mpc]']
dpsi_2 = psi_2*(EYMH(1,-34.5388)['y_eymh']*EYMH(1,-34.5388)['xi_eymh']/EYMH(1,-34.5388)['H [1/Mpc]'] -1 -Sigma(1,-34.5388))
Phi    = np.sqrt(3)*EYMH(1,-34.5388)['w_eymh']/EYMH(1,-34.5388)['s_eymh']**2/EYMH(1,-34.5388)['xi_eymh']
dPhi   = np.sqrt(6)*EYMH(1,-34.5388)['z_eymh']/EYMH(1,-34.5388)['H [1/Mpc]']
```

```python
def avf(l):
    # create instance of the class "Class"
    ani_vf = Class()
    # pass input parameters
    ani_vf.set({'omega_b':0.0223828,'omega_cdm':0.1201075,'h':0.67810,'A_s':2.100549e-09,'n_s':0.9660499,'tau_reio':0.05430842,'Omega_fld':0.,'Omega_scf':0.,'Omega_anq':0.,'Omega_sde':0.,'Omega_a2f':0,'Omega_eymh':0,'Omega_avf':1.,'avf_parameters_1':l})
    #LambdaCDM.set({'output':'tCl,pCl,lCl,mPk','lensing':'yes','P_k_max_1/Mpc':3.0})
    # run class
    ani_vf.compute()
    return pd.DataFrame(ani_vf.get_background())
```

```python
z = avf(0.01)['z']
H = avf(0.01)['H [1/Mpc]']


Omega_r   = (avf(0.01)['(.)rho_g']+avf(0.01)['(.)rho_ur']) /H**2
Omega_b   = (avf(0.01)['(.)rho_b'])/H**2
Omega_cdm = (avf(0.01)['(.)rho_cdm'])/H**2
Omega_DE  = (avf(0.01)['(.)rho_avf'])/H**2
w_DE      = (avf(0.01)['(.)p_avf'])/avf(0.01)['(.)rho_avf']



z_avf = avf(0.01)['z_avf']
v_avf = avf(0.01)['v_avf']
x_avf = avf(0.01)['x_avf']

def Sigma(l):
  return avf(l)['sigma_avf']/avf(l)['H [1/Mpc]']

def w_DE(l):
  return avf(l)['(.)p_avf']/avf(l)['(.)rho_avf']

q = -1. + 0.01*v_avf*v_avf*x_avf*x_avf/2./H + 2.*z_avf*z_avf/H/H + 3.*(Omega_b+Omega_cdm)/2. + 2.*Omega_r + 3.*Sigma(0.01)**2 
w_eff = -1. + 2.*(1.+q)/(1-Sigma(0.01)**2)/3 - 2.*Sigma(0.01)**2/(1-Sigma(0.01)**2)
```
